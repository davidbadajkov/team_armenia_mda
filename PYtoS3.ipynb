```python
# Upload a new file from colab to AWS S3 bucket
s3.Bucket('s3grouparmenia').upload_file(Filename='1405transcript_preprocessed_compact.csv', Key='1405transcript_preprocessed_compact.csv')
```

```python
#read file from S3 bucket to colab by using boto3

s3_file_key = '1405transcript_preprocessed_compact.csv'
bucket = 's3grouparmenia'

s3 = boto3.client('s3')
obj = s3.get_object(Bucket=bucket, Key=s3_file_key)

initial_df = pd.read_csv(io.BytesIO(obj['Body'].read()))
```
