## Consolidating all transcripts into one dataframe

```python
path = '/Users/danielbejarano/MDA/Text files 1970-2018'
os.chdir(path)
```

```python
os.listdir(path)
```

```python
""" I'm gonna create some empty lists to represent important features of the transcripts, and 
I will just get the values from the file names. Then I'm gonna iteratively add each transcript to the lists"""
years = []
countries = []
transcripts = []
sessions = []
```

```python
folders = os.listdir(path)
for root, dirs, files in os.walk(path):
    for name in files:
        if name.endswith('.txt'):
            with open(os.path.join(root, name), 'r') as f:
                data = f.read()
                try:
                    transcripts.append(data)
                except:
                    transcripts.append(np.nan)
            countries.append(name[:3])
            years.append(name[7:11])
            sessions.append(name[4:6])
```

```python
dic = {
    'Year': years,
    'Session': sessions,
    'Country': countries, 
    'Transcript': transcripts
}

df1 = pd.DataFrame(dic)
df1.head()
```

```python
df1.to_csv('consolidated-transcripts', encoding = 'utf-8')
```
