```python
import nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from glob import glob
import os
%matplotlib inline
```

```python
path = 'C:/Users/david/Documents/0_kul/1_ma_stds_1/0_mda/project/data/0_data/Converted sessions'
os.chdir(path)
```

```python

""" I'm gonna create some empty lists to represent important features of the transcripts, and 
I will just get the values from the file names. Then I'm gonna iteratively add each transcript to the lists"""
years = []
countries = []
transcripts = []
sessions = []
```

```python
folders = os.listdir(path)
for root, dirs, files in os.walk(path):
    for name in files:
        if name.endswith('.txt'):
            with open(os.path.join(root, name), 'r', encoding='utf-8') as f:
                data = f.read()
                try:
                    transcripts.append(data)
                except:
                    transcripts.append(np.nan)
            countries.append(name.split('_')[0])
            years.append(name.split('_')[-1])
            sessions.append(name.split('_')[1])
```

```python
years_cleaned = []

for year in years:
    years_cleaned.append(year.replace('.txt', ''))
```

```python

dic = {
    'Year': years_cleaned,
    'Session': sessions,
    'Country': countries, 
    'Transcript': transcripts
}

df = pd.DataFrame(dic)
df.head(-5)
```
